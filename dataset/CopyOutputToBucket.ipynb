```python
from google.cloud import storage
import os
from pathlib import Path

GOOGLE_APPLICATION_CREDENTIALS="/home/fenics/shared/bucket-key.json"
client = storage.Client.from_service_account_json(GOOGLE_APPLICATION_CREDENTIALS)
bucket = client.get_bucket('learning-amg-ftetwild-matrices')
```

```python
basepath='/home/fenics/shared/output'
```

```python
def walk_outputs(max_files=10):
    count = 0
    for file in os.listdir("output"):
        if "." in file:
            continue
        
        if count > max_files:
            break
        filename = "{}".format(file)
        yield filename
        count += 1
```

```python
import concurrent.futures

def do_problem(problem):
    files = ["{}-pressure_A.npz", "{}-velocity_A.npz", "{}-velocity_correction_A.npz", "{}.json",
    "{}-pressure_b.npz", "{}-velocity_b.npz", "{}-velocity_correction_b.npz"]

    for filename in files:
        blob = bucket.blob('{}/{}'.format(problem, filename.format(problem)))
        blob.upload_from_filename(filename='{}/{}/{}'.format(basepath, problem, filename.format(problem)))
        
    return True

with concurrent.futures.ThreadPoolExecutor(max_workers=16) as executor:
    future_to_problem = {executor.submit(do_problem, problem): problem for problem in walk_outputs(10)}
    for future in concurrent.futures.as_completed(future_to_problem):
        problem = future_to_problem[future]
        try:
            data = future.result()
        except Exception as exc:
            print("'%r generated an exception: %s' % (problem, exc)")
        else:
            print("{} succeeded".format(problem))
```
