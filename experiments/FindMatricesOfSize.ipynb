```python
# We want to find matrices of size 0...max
import numpy as np
from matplotlib import pyplot as plt 
import os
from pathlib import Path
import json
```

```python
data = {}

def walk_outputs(max_files=10):
    count = 0
    for file in os.listdir("output"):
        if "." in file:
            continue
        
        if count > max_files:
            break
        filename = "output/{}/{}.json".format(file, file)
        if Path(filename).is_file():
            file = file.split('.')[0]
            count += 1
            yield file, filename 
               
        
for name, file in walk_outputs(10000):
    with open(file, 'r') as json_file:
        this_data = json.load(json_file)
        data[name] = this_data
```

```python
result = [(0,0)] * len(data)
for i, (k,v) in enumerate(data.items()):
    m = [x for x in v['matrices'] if x['name'] == 'velocity_A'][0]
    result[i] =  v['problem'], m['shape'][0]
result = sorted(result, key=lambda x: x[1])
```
